```python
import os
import numpy as np
import pandas as pd
from scipy.signal import savgol_filter
from pykrx import stock
import matplotlib.pyplot as plt
import time
plt.rcParams['font.family'] = 'Malgun Gothic'
from KorDivStockPricer import KorDivStockPricer
```

```python
def makeDirIfNotExist(dir):
    try:
        if not os.path.exists(dir):
            os.makedirs(dir)
    except Exception as e:
        print(str(e))
```

```python
OUT_DIR = os.path.join(os.getcwd(), 'output')
OUT_DIR
```

```python
makeDirIfNotExist(OUT_DIR)
```

```python
START_YEAR = 2012
END_YEAR = 2021
```

```python
df = pd.read_excel('./data/tiger_divgrowth50_div_analyze_result.xlsx')
df.head(3)
```

```python
df.tail(3)
```

```python
#df_target = df.head(2)
df_target = df
pricingResults = []
for code in df_target.종목코드:
    code = str(code).zfill(6)
    print(f'code = {code}')
    stockPricer = KorDivStockPricer(code)
    stockPricer.getStockName()
    stockPricer.collectStockData()
    pricingResult = stockPricer.doStockPricing()
    pricingResults.append(pricingResult)
    stockPricer.savePriceDivChart()
    time.sleep(1)
```

```python
len(pricingResults)
```

```python
df_total = None
for pricingResult in pricingResults:
    #print(pricingResult)
    df_pricing_result = pd.DataFrame(pricingResult)
    if df_total is None:
        df_total = df_pricing_result
    else:
        df_total = pd.concat([df_total, df_pricing_result])
```

```python
df_total = df_total.reset_index()
```

```python
df_total.head(3)
```

```python
df_total.tail(3)
```

```python
df_final = pd.concat([df, df_total], axis=1)
df_final
```

```python
df_final.drop('index', inplace=True, axis=1)
df_final.head(3)
```

```python
resultFilePath = os.path.join(OUT_DIR, 'tiger_divgrowth50_pricing_result.xlsx')
df_final.to_excel(resultFilePath, index=False)
```

```python
df_pricing_result = pd.read_excel(resultFilePath)
```
