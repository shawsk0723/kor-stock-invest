```python
import pandas as pd
from pykrx import stock
```

```python
target_date = '20230213'
```

```python
"""
코스피 상장 주식 펀더멘탈 정보 읽어 오기
'"""
df_f = stock.get_market_fundamental_by_ticker(target_date, market='KOSPI')
```

```python
df_f
```

```python
"""
PER 0 이하 제거
"""
df_f.drop(df_f[df_f.PER <= 0].index, inplace=True)
```

```python
df_f.drop(df_f[df_f.PBR <= 0].index, inplace=True)
```

```python
df_f
```

```python
"""
통계 정보 보기
"""
df_f.describe()
```

```python
df = df_f
```

```python
df['PER*PBR'] = df.PER * df.PBR
```

```python
df
```

```python
df['PER_PBR_RANK'] = df['PER*PBR'].rank()
```

```python
df
```

```python
df.sort_values(by='PER_PBR_RANK', inplace=True)
```

```python
df
```

```python
"""
이름 추가하기
"""
names = []
for ticker in df.index:
    try:
        name = stock.get_market_ticker_name(ticker)
        names.append(name)
    except Exception as e:
        print(f' exception happens while getting name of ticker = {ticker}')
        print(e)
        names.append('N/A')
```

```python
df.insert(0, 'NAME', names)
```
